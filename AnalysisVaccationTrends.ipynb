```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
poland_df = pd.read_csv("datasets/Poland_Vaccation.csv", parse_dates=['Miesiąc'], date_format="%Y-%m")
usa_df = pd.read_csv("datasets/USA_Vaccation.csv", parse_dates=['Miesiąc'], date_format="%Y-%m")
uk_df = pd.read_csv("datasets/UK_Vaccation.csv", parse_dates=['Miesiąc'], date_format="%Y-%m")
```

```python
poland_df["date"] = poland_df["Miesiąc"]
usa_df["date"] = usa_df["Miesiąc"]
uk_df["date"] = uk_df["Miesiąc"]
poland_df.drop('Miesiąc', axis=1, inplace=True)
usa_df.drop('Miesiąc', axis=1, inplace=True)
uk_df.drop('Miesiąc', axis=1, inplace=True)
poland_df.head()
usa_df.head()
uk_df.head()
```

```python
poland_df.set_index("date", inplace=True)
usa_df.set_index("date", inplace=True)
uk_df.set_index("date", inplace=True)
```

```python
poland_df.rename(columns={'Vacation: (Polska)': 'PL'}, inplace=True)
usa_df.rename(columns={'Vacation: (Stany Zjednoczone)': 'US'}, inplace=True)
uk_df.rename(columns={'Vacation: (Wielka Brytania)': 'UK'}, inplace=True)
```

```python
combined_df = pd.concat([poland_df, usa_df, uk_df], axis=1)
combined_df.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(combined_df.index, combined_df['PL'], label='PL')
plt.plot(combined_df.index, combined_df['US'], label='US')
plt.plot(combined_df.index, combined_df['UK'], label='UK')
plt.xlabel('Date')
plt.ylabel('Vacation')
plt.title('Vacation Trends')
plt.legend()
plt.show()
```

```python
combined_df.describe()
```

```python
plt.figure(figsize=(10,6))
plt.hist(combined_df['PL'].dropna(), alpha=0.5, label='PL')
plt.hist(combined_df['US'].dropna(), alpha=0.5, label='US')
plt.hist(combined_df['UK'].dropna(), alpha=0.5, label='UK')
plt.xlabel('Vacation')
plt.ylabel('Frequency')
plt.title('Vacation Trends')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.kdeplot(combined_df['PL'].dropna(), label='PL')
sns.kdeplot(combined_df['US'].dropna(), label='US')
sns.kdeplot(combined_df['UK'].dropna(), label='UK')
plt.xlabel('Vacation')
plt.ylabel('Density')
plt.title('Vacation Trends')
plt.legend()
plt.show()
```
